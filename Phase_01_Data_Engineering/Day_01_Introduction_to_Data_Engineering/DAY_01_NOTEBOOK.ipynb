```python
import pandas as pd

deleted_indices = df.sample(n=5).index

deleted_rows = df.loc[deleted_indices]

df = df.drop(index=deleted_indices)

print(f"Deleted index: {deleted_indices.tolist()}")
print("\nDeleted Rows:")
display(deleted_rows)

print(f"\n new rows: {len(df)}")
```
